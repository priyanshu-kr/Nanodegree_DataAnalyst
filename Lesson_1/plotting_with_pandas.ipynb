```python
import pandas as pd
```

```python
df_census = pd.read_csv('data/census_income_data.csv')
df_census.info()
```

```python
df_census.hist()
```

```python
df_census.hist(figsize = (12, 12));
# use semicolon to hide the written extra data
```

```python
# we can also use hist() function on a specific column
df_census['age'].hist(figsize = (10, 6));
```

```python
# instead of using hist() function, we can use plot() function and provide the kind of chart we need
df_census['age'].plot(kind='hist', figsize=(10,6));
```

```python
# value_counts() function aggregates counts for each unique value in a column
df_census['education'].value_counts()
```

```python
# we can use plot() funtion to get the bar chart of education column
df_census['education'].value_counts().plot(kind='bar');
```

```python
# similarly we also need value_counts for creating pie charts
df_census['workclass'].value_counts().plot(kind='pie', figsize=(8, 8));
```

```python
#################################################################################################
# let's get to the cancer dataset to practice plotting scatter plots and box plots

df_cancer = pd.read_csv('data/cancer_data_edited.csv')
df_cancer.info()
```

```python
# this helps to look into the relationship among numerical variables with scatter plots
pd.plotting.scatter_matrix(df_cancer, figsize=(15, 15));
```

```python
# we can also get a single plot using this...
df_cancer.plot(x='compactness', y='concavity', kind='scatter');
```

```python
# plotting box plot for specific column
df_cancer['concave_points'].plot(kind='box');
```
