```python
import pandas as pd
import numpy as np
```

```python
# `sep` for separator, to split file, default: comma
df = pd.read_csv('data/coffee_dataset.csv', sep=',') 
df.head(10)
```

```python
# changing the header of the file. It will cut off everything above it.
df = pd.read_csv('data/coffee_dataset.csv', header=2) 
df.head(10)
```

```python
# If columns labels are not included in your file, 
# you can use header=None to prevent your first line of data from being misinterpreted as column labels.
df = pd.read_csv('data/coffee_dataset.csv', header=None) 
df.head(10)
```

```python
# we can specify our own column names
labels = ['ID', 'AGE', 'Drinks_Coffee', 'Height']
df = pd.read_csv('data/coffee_dataset.csv', names=labels) 
df.head()
```

```python
labels = ['ID', 'AGE', 'Drinks_Coffee', 'Height']
df = pd.read_csv('data/coffee_dataset.csv', names=labels, header=0) 
df.head()
```

```python
# we can make specify one or more of our columns to be the index of our dataframe
df = pd.read_csv('data/coffee_dataset.csv', index_col='user_id') 
df.head()
```

```python
df = pd.read_csv('data/coffee_dataset.csv', index_col=['user_id', 'height']) 
df.head()
```

```python
# We can use 'loc' and 'iloc' to select data from the dataframe;
# 'loc' uses labels of rows and columns to select data, and
# 'iloc' uses index numbers.
```

```python
# select all the columns from house_id to the price column using names;
d1_means = d1.loc[:, 'house_id':'price']
d1_means.head()
```

```python
# selecting all the columns from house_id to price column using index numbers;
d1_means = d1.iloc[:,:7]
d1_means.head()
```

```python
# saving to the csv file for further use
d1_means.to_csv('data/house_prices.csv', index=False)
```

```python
d1_means = d1.iloc[:, np.r_[0:3,4,5:6]]
d1_means.head()
```
