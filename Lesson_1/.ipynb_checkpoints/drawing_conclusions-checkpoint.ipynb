```python
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv('data/cancer_data_edited.csv')
df.head(22).style.background_gradient(cmap='Oranges')
```

```python
# we can separate out the Malignant tumor dataframe using mask.
df_m = df[df['diagnosis'] == 'M']
df_m.head(5).style.background_gradient(cmap='Oranges')
```

```python
# mask is used in the above query
mask = df['diagnosis'] == 'M' 
df_m = df[mask]
df_m
```

```python
df_m['area'].describe()
```

```python
# now for Benign tumors
df_b = df[df['diagnosis'] == 'B']
df_b['area'].describe()
```

```python
print('The mean area of malignant tumors is {0:.4f} \
while that of benign tumors is {0:.4f}.'.format(df_m['area'].mean(), df_b['area'].mean()))
```
