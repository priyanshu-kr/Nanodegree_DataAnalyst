```python
import numpy as np
import pandas as pd
```

```python
red_df = pd.read_csv('winequality-red.csv', sep=';')
white_df = pd.read_csv('winequality-white.csv', sep=';')
```

```python
red_df.info()
```

```python
white_df.info()
```

```python
red_df.shape
```

```python
white_df.shape
```

```python
color_red = np.repeat('Red', 1599, axis=0)
color_red
```

```python
color_white = np.repeat('White', 4898, axis=0)
color_white
```

```python
red_df['color'] = color_red
red_df.tail()
```

```python
white_df['color'] = color_white
white_df.tail()
```

```python
wine_df = red_df.append(white_df, sort=False)
wine_df.head()
```

```python
wine_df.to_csv('winequality-edited.csv', index=False)
```
