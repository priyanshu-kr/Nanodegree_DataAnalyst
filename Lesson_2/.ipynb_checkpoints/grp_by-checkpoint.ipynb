```python
import pandas as pd
```

```python
wine = pd.read_csv('winequality-edited.csv')
wine.head()
```

```python
wine.mean()
```

```python
# to get the mean grouped by any specific columns
wine.groupby('quality').mean()
```

```python
wine.groupby(['quality', 'color'], as_index=True).mean()
```

```python
wine.groupby(['quality', 'color'], as_index=False)['pH', 'density'].mean()
```

```python
# pH is a quantitative variable without clear categories. However, there is a simple fix to this.
# You can create a categorical variable from a quantitative variable by creating your own categories.
# pandas' cut function let's you "cut" data in groups.
```

```python
wine.groupby('color').mean().quality
```

```python
wine.describe().pH
```

```python
bin_edges = [2.72, 3.11, 3.21, 3.32, 4.01]
bin_names = ['high', 'mod_high', 'medium', 'low']
```

```python
wine['acidity_levels'] = pd.cut(wine['pH'], bin_edges, labels=bin_names)
wine.head()
```

```python
wine.groupby('acidity_levels').mean().quality
```

```python
wine.to_csv('winequality-edited.csv', index=False)
```
